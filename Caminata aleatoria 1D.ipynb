## Problema

Tomemos un punto que se pueda mover a la derecha con probabilidad $p$  
y a la izquierda con probabilidad $q$, tal que:

$$
p + q = 1
$$

En cada paso el movimiento es independiente de los anteriores.

---

## Planteamiento del Código

La lógica que se sigue para construir el programa es la siguiente:

Primero, se le pregunta al usuario si desea introducir una probabilidad específica únicamente para $p$.

- Si el usuario introduce un valor, ese será $p$.
- Si el usuario no introduce ningún número, el sistema genera un valor aleatorio entre 0 y 1 para $p$.

Una vez determinado $p$, se calcula automáticamente:

$$
q = 1 - p
$$

Se fija de manera estándar el número de pasos como:

$$
n = 100
$$

Después, en cada uno de los 100 pasos:

- Se genera un número aleatorio entre 0 y 1.
- Si el número es menor o igual que $p$, el punto se mueve a la derecha.
- En caso contrario, se mueve a la izquierda.

Se acumula la posición en un contador y se guardan todos los valores para poder graficar la trayectoria completa.

---

## Código en Python

```python
import random
import matplotlib.pyplot as plt

def derecha_izquierda():
    
    n = 100
    
    entrada = input("Introduce la probabilidad p (Enter para aleatoria): ")
    
    if entrada == "":
        p = random.uniform(0,1)
    else:
        p = float(entrada)
    
    q = 1 - p
    
    print("p =", p)
    print("q =", q)
    
    contador = 0
    valores = []
    
    for i in range(1, n+1):
        
        prob = random.uniform(0,1)
        
        if prob <= p:
            x = 1
        else:
            x = 0
        
        if x == 1:
            contador += 1
        else:
            contador -= 1
        
        valores.append(contador)
    
    return valores


resultado = derecha_izquierda()

plt.plot(resultado)
plt.xlabel("Paso n")
plt.ylabel("Posición")
plt.title("Caminata Aleatoria Simple (n=100)")
plt.axhline(0)
plt.show()
```

---

## Análisis del Sistema

### 1. ¿El punto regresa al origen?

La respuesta depende completamente del número de eventos y de las probabilidades.

Si $p$ y $q$ son muy diferentes, el punto tenderá a desplazarse hacia el lado con mayor probabilidad. En ese caso, la posibilidad de regresar al origen es muy baja.

Cuando la diferencia entre $p$ y $q$ es pequeña, el comportamiento es más equilibrado. El punto oscila más y la probabilidad de retorno aumenta.

Mientras más similares sean $p$ y $q$, mayor es la posibilidad de que el punto regrese al origen.

---

### 2. ¿Cuánto se aleja en promedio del origen?

El alejamiento promedio depende directamente de las probabilidades.

Si $p > q$, existe una tendencia hacia la derecha.  
Si $p < q$, existe una tendencia hacia la izquierda.  
Si $p = q$, no hay preferencia direccional.

Matemáticamente:

$$
\mathbb{E}[S_n] = n(p-q)
$$

Esto muestra que la diferencia entre probabilidades determina la tendencia promedio del sistema.

---

### 3. ¿Cuál es el alejamiento máximo?

El alejamiento máximo posible después de $n$ pasos ocurre cuando todos los movimientos son en la misma dirección:

$$
|S_n|_{\text{máx}} = n
$$

La probabilidad de que esto ocurra es $p^{100}$ o $q^{100}$, lo cual se vuelve muy pequeño cuando $n$ es grande.

---

## Función de Probabilidad

En $n$ pasos, el número de movimientos hacia la derecha sigue una distribución binomial:

$$
N_n \sim \text{Binomial}(n,p)
$$

Si hubo $k$ movimientos hacia la derecha, la posición final es:

$$
S_n = 2k - n
$$

Por lo tanto,

$$
P(S_n = s) =
\binom{n}{k} p^k q^{n-k}
$$

donde

$$
k = \frac{n+s}{2}
$$

siempre que $n+s$ sea un número par.
