# Distribución Weibull

La distribución Weibull se utiliza en confiabilidad y análisis de tiempos de falla.
Usando el método de la transformada inversa y la función acumulada:

$$
F(x)=1-e^{-x^\alpha}
$$

se obtiene la expresión:

$$
X = (-\ln(U))^{1/\alpha}
$$

donde:

$$
U \sim U(0,1)
$$

```python
import numpy as np

# Función para generar Weibull
def weibull_inversa(alpha, n):
    
    U = np.random.uniform(0,1,n)
    
    X = (-np.log(U))**(1/alpha)
    
    return X

# Parámetros
alpha = 2
n = 10000

# Generar muestra
muestra = weibull_inversa(alpha, n)

# Imprimir algunos valores
print("Primeros 10 valores generados:\n")
print(muestra[:10])
```

```python
import matplotlib.pyplot as plt

plt.hist(muestra, bins=50, density=True)

plt.title("Histograma - Weibull")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.show()
```

```python
print("Media:", np.mean(muestra))

print("Varianza:", np.var(muestra))

print("Desviación estándar:", np.std(muestra))
```

# Distribución Gaussiana Estándar

La distribución normal estándar es una de las distribuciones más importantes en probabilidad y estadística. 
Su función de distribución acumulada está definida por:

$$
\Phi(x)=\int_{-\infty}^{x}\frac{1}{\sqrt{2\pi}}e^{-t^2/2}dt
$$

Aplicando el método de transformada inversa:

$$
\Phi(x)=U
$$

donde:

$$
U \sim U(0,1)
$$

Sin embargo, la función acumulada normal no posee una inversa analítica cerrada, debido a que la integral que la define no puede expresarse mediante funciones elementales. Por esta razón, la inversa se calcula mediante métodos numéricos computacionales:

$$
X = \Phi^{-1}(U)
$$

```python
import numpy as np
from scipy.stats import norm

# Función para generar Normal estándar
def normal_inversa(n):
    
    U = np.random.uniform(0,1,n)
    
    X = norm.ppf(U)
    
    return X

# Parámetros
n = 10000

# Generar muestra
muestra = normal_inversa(n)

# Imprimir algunos valores
print("Primeros 10 valores generados:\n")
print(muestra[:10])
```

```python
import matplotlib.pyplot as plt

plt.hist(muestra, bins=50, density=True)

plt.title("Histograma - Normal Estándar")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.show()
```

```python
print("Media:", np.mean(muestra))

print("Varianza:", np.var(muestra))

print("Desviación estándar:", np.std(muestra))
```

# Distribución Cauchy Estándar

La distribución Cauchy se utiliza en algunos modelos físicos y fenómenos con colas pesadas. 
A diferencia de la distribución normal, la distribución Cauchy no posee media ni varianza definidas.

Su función de distribución acumulada es:

$$
F(x)=\frac{1}{\pi}\arctan(x)+\frac12
$$

Aplicando el método de transformada inversa:

$$
U=\frac{1}{\pi}\arctan(x)+\frac12
$$

Despejando:

$$
X=\tan\left(\pi(U-\frac12)\right)
$$
donde:

$$
U\sim U(0,1)
$$

```python
import numpy as np

# Función para generar Cauchy estándar
def cauchy_inversa(n):
    
    U = np.random.uniform(0,1,n)
    
    X = np.tan(np.pi*(U - 0.5))
    
    return X

# Parámetros
n = 10000

# Generar muestra
muestra = cauchy_inversa(n)

# Imprimir algunos valores
print("Primeros 10 valores generados:\n")
print(muestra[:10])
```

```python
import matplotlib.pyplot as plt

plt.hist(muestra, bins=50, density=True, range=(-10,10))

plt.title("Histograma - Cauchy Estándar")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.show()
```

```python
print("Media:", np.mean(muestra))

print("Varianza:", np.var(muestra))

print("Desviación estándar:", np.std(muestra))
```

# Distribución Logística Estándar

La distribución logística es similar a la distribución normal, aunque posee colas ligeramente más pesadas. 
Es utilizada en modelos logísticos y en regresión logística.

Su función de distribución acumulada es:

$$
F(x)=\frac{1}{1+e^{-x}}
$$

Aplicando el método de transformada inversa:

$$
U=\frac{1}{1+e^{-x}}
$$

Despejando:

$$
X=\ln\left(\frac{U}{1-U}\right)
$$

donde:

$$
U\sim U(0,1)
$$

```python
import numpy as np

# Función para generar Logística estándar
def logistica_inversa(n):
    
    U = np.random.uniform(0,1,n)
    
    X = np.log(U/(1-U))
    
    return X

# Parámetros
n = 10000

# Generar muestra
muestra = logistica_inversa(n)

# Imprimir algunos valores
print("Primeros 10 valores generados:\n")
print(muestra[:10])
```

```python
import matplotlib.pyplot as plt

plt.hist(muestra, bins=50, density=True)

plt.title("Histograma - Logística Estándar")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.show()
```

```python
print("Media:", np.mean(muestra))

print("Varianza:", np.var(muestra))

print("Desviación estándar:", np.std(muestra))
```

# Distribución Pareto

La distribución Pareto se utiliza para modelar fenómenos con colas pesadas, como distribución de riqueza, ingresos y tamaños de empresas. 
Se caracteriza por la presencia de eventos extremos con mayor probabilidad que en otras distribuciones.

Su función de distribución acumulada es:

$$
F(x)=1-x^{-\alpha}, \quad x\ge1
$$

Aplicando el método de transformada inversa:

$$
U=1-x^{-\alpha}
$$

Despejando:

$$
X=(1-U)^{-1/\alpha}
$$

Como:

$$
1-U \sim U(0,1)
$$

también puede escribirse como:

$$
X=U^{-1/\alpha}
$$

```python
import numpy as np

# Función para generar Pareto
def pareto_inversa(alpha, n):
    
    U = np.random.uniform(0,1,n)
    
    X = U**(-1/alpha)
    
    return X

# Parámetros
alpha = 3
n = 10000

# Generar muestra
muestra = pareto_inversa(alpha, n)

# Imprimir algunos valores
print("Primeros 10 valores generados:\n")
print(muestra[:10])
```

```python
import matplotlib.pyplot as plt

plt.hist(muestra, bins=50, density=True, range=(1,10))

plt.title("Histograma - Pareto")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.show()
```

```python
print("Media:", np.mean(muestra))

print("Varianza:", np.var(muestra))

print("Desviación estándar:", np.std(muestra))
```
